```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("car data.csv")

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Bar Graph: Average Selling Price by Fuel Type
plt.figure(figsize=(6, 3))
sns.barplot(x='Fuel_Type', y='Selling_Price', data=df, estimator=np.mean, palette='viridis')
plt.title('Average Selling Price by Fuel Type')
plt.ylabel('Average Selling Price (in Lakhs)')
plt.xlabel('Fuel Type')
plt.tight_layout()
plt.show()

# 2. Histogram: Distribution of Present Price
plt.figure(figsize=(6, 3))
plt.hist(df['Present_Price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Present Price')
plt.xlabel('Present Price (in Lakhs)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Pie Chart: Seller Type Distribution
seller_counts = df['Seller_Type'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(seller_counts, labels=seller_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Seller Type Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
